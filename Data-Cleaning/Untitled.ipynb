```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df
```

```python
#Identifies missing values
df.isnull()
```

```python
#Summing over the columns gives the total count of missing values for each column.
df.isnull().sum()
```

```python
#If you want to drop rows with missing values in specific columns only, use the subset argument.
```

```python
df_cleaned = df.dropna(subset=['Age','Cabin','Embarked'])
print(df_cleaned)
```

```python
# Fill missing values in the DataFrame with the global constant
```

```python
df_filled = df.fillna("Unknown")
print(df_filled)
```

```python
#Mean/Median/Mode Imputation
```

```python
mean_imputation = df['Age'].fillna(df['Age'].mean())
print(mean_imputation)
```

```python
#For categorical or string data, you can use mode()

median_imputation = df['Cabin'].fillna(df['Cabin'].mode()[0])
print(median_imputation)

#print("mode = " + df['Cabin'].mode())
#print("mode = " + df['Cabin'].mode()[0])
```

```python
mode_imputation = df['Embarked'].fillna(df['Embarked'].mode())
print(mode_imputation)

#print("mode = " + df['Embarked'].mode())
```

```python
#Forward and Backward Fill
```

```python
forward_fill = df['Age'].ffill()
print(forward_fill)
```

```python
backward_fill = df['Cabin'].bfill()
print(backward_fill)
```

```python
backward_fill = df['Embarked'].bfill()
print(backward_fill)
```

```python
#Interpolation Techniques
#Interpolation is typically applied to numeric data in a dataset, especially when you have missing values (NaN) that need to be estimated.
#you need to install the scipy library.
```

```python
linear_interpolation = df['Age'].interpolate(method='linear')
print(linear_interpolation)
```

```python
polynomial_interpolation = df['Age'].interpolate(method='polynomial', order=2)
print(polynomial_interpolation)
```

```python
spline_interpolation = df['Age'].interpolate(method='spline', order=3)
print(spline_interpolation)
```

```python
nearest_interpolation = df['Age'].interpolate(method='nearest')
print(nearest_interpolation)
```
